```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import re
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
from google.colab import files
uploaded = files.upload()
df = pd.read_csv("/content/Synthetic_YouTube_Trending_Data.csv")
df.head()
```

```python
df['title'] = df['title'].astype(str).apply(lambda x: re.sub(r'[^a-zA-Z0-9 ]', '', x))
df['tags'] = df['tags'].astype(str).apply(lambda x: re.sub(r'[^a-zA-Z0-9|]', '', x))
```

```python
df['publish_time'] = pd.to_datetime(df['publish_time'])
df['trending_date'] = pd.to_datetime(df['trending_date'])
```

```python
sid = SentimentIntensityAnalyzer()
df['sentiment_score'] = df['title'].apply(lambda x: sid.polarity_scores(x)['compound'])
df['sentiment_label'] = df['sentiment_score'].apply(lambda score: 'Positive' if score > 0.05 else ('Negative' if score < -0.05 else 'Neutral'))
```

```python
fig1 = px.box(df, x='category', y='views', color='category', title='Views by Category')
fig1.show()
```

```python
fig2 = px.scatter(df, x='likes', y='dislikes', color='category', hover_data=['title'], title='Likes vs Dislikes by Category')
fig2.show()
```

```python
fig3 = px.histogram(df, x='sentiment_label', color='sentiment_label', title='Sentiment Distribution of Video Titles')
fig3.show()
```

```python
fig4 = px.histogram(df, x='country', color='category', title='Trending Video Count by Country and Category')
fig4.show()
```

```python
df.to_csv("Cleaned_YouTube_Trending_Data.csv", index=False)
print("Cleaned dataset saved as 'Cleaned_YouTube_Trending_Data.csv'")
```
